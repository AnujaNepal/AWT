```python
import numpy as n
import numpy as np
num_data = np.array([12,15,14,10,18,20,22,19,17,16,14,13,11])
```

# Histogram




```python
#Show the frequency of value grouped into bins;approximates the empirical PDF.
import matplotlib.pyplot as plt

plt.hist(num_data,bins=5,density=True)
plt.title("Histogram - Numerical Data")
plt.xlable("value")
plt.show()
```

```python
# Emprical CDF
# Description:show cumulative probability p(X<x)
```

```python
x= np.snort(num_data)
y= np.arange(1, len(num_data)+1) / len(num_data)

plt.plot(x,y)
plt,title("ECDF - Numerical Data")
plt.
```

```python
from scipy.stats import gaussian_kde

kde = gaussian_kde(num_data)
x_vals = np.linspace(min(num_data),max(num_data),100)
y_vals = kde(x_vals)

plt.plot(x_vals , y_vals)
plt.title("KDE - Numerical Data")
plt.xlabel("value")
plt.ylabel("Density")
plt.show()
```

# Box Plot
# Shows median, quartiles and outliers

```python
plt.boxplot(num_data)
plt.title("Box Plot")
plt.show()
```

```python
plt.violinplot(num_data)
plt.title("Violin plot")
plt.show()
```
