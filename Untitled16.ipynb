```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
```

```python
# Load dataset
iris = load_iris()
X = iris.data  # features
y = iris.target  # actual labels (not used in clustering)

# Standardize the features (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Reduce dimensions to 2 for plotting
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
def plot_clusters(X, labels, title):
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette="Set1", s=60)
    plt.title(title)
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)
plot_clusters(X_pca, kmeans_labels, "K-Means Clustering")
```

```python
dbscan = DBSCAN(eps=0.6, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)
plot_clusters(X_pca, dbscan_labels, "DBSCAN Clustering")
```

```python
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(X_scaled)
plot_clusters(X_pca, hierarchical_labels, "Hierarchical Clustering")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
# Create synthetic 2D moon-shaped data
X, _ = make_moons(n_samples=300, noise=0.07, random_state=42)

# Visualize the raw data
plt.figure(figsize=(6, 4))
plt.scatter(X[:, 0], X[:, 1], s=30, color='blue')
plt.title("Original Moons Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# DBSCAN parameters: eps and min_samples
dbscan = DBSCAN(eps=0.3, min_samples=5)
db_labels = dbscan.fit_predict(X_scaled)
```

```python
# Plot clustered data
plt.figure(figsize=(6, 4))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=db_labels, palette="Set1", s=60)
plt.title("DBSCAN on Moons Dataset")
plt.xlabel("Standardized Feature 1")
plt.ylabel("Standardized Feature 2")
plt.legend(title="Cluster")
plt.show()
```
